# hsv image equalize

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import skimage 
import numpy as np
```

```python
im = skimage.data.coffee()
im
```

```python
plt.imshow(im)
```

```python
im1 = np.array([[[255.,0.,0.]]])/255
skimage.color.rgb2hsv(im1)
```

```python
skimage.color.hsv2rgb(np.array([[[0,1.,1.]]]))
```

```python
_=plt.hist(im[:,:,0].ravel(),color='r',alpha=0.5,bins=50,density=True)
_=plt.hist(im[:,:,1].ravel(),color='g',alpha=0.5,bins=50,density=True)
_=plt.hist(im[:,:,2].ravel(),color='b',alpha=0.5,bins=50,density=True)
```

```python
hsv = skimage.color.rgb2hsv(im)
plt.figure(figsize=(12,6))
_=plt.hist(hsv[:,:,0].ravel(),color='r',alpha=0.5,bins=250,density=True)
_=plt.hist(hsv[:,:,1].ravel(),color='g',alpha=0.5,bins=250,density=True)
_=plt.hist(hsv[:,:,2].ravel(),color='b',alpha=0.5,bins=250,density=True)
```
