히스토그램은 이미지를 구성하는 픽셀값 분포에 대한 그래프입니다. X축은 픽셀값으로 범위는 0 ~ 255 사이입니다. Y축은 이미지에서 해당 픽셀값을 가진 픽셀의 개수입니다. 

```python
def linearmap(x , xmin , xmax , tmin=0 , tmax=255):
    if(x < xmin):
        return tmin
    if(x > xmax):
        return tmax
    t = ((x-xmin) / (xmax-xmin)) * (tmax-tmin) + tmin
    return t
    
```

```python
def hstretch(im, xmin, xmax, tmin , tmax):
    j = np.empty_like(im).ravel()
    ii = im.ravel()
    for i in range(j.shape[0]):
        j[i] = linearmap(ii[i],xmin,xmax,tmin,tmax)
    
    return j.reshape(im.shape)
```

```python
def cumulative(count):
    cdf = np.zeros(256)
    cdf[0] = count[0]
    for i in range(1,256):
        cdf[i] = cdf[i-1] + count[i]
    return cdf
```

```python
N = im.shape[0] * im.shape[1]
cul=cumulative(count)/N
count = count/N
```

```python

_=plt.plot(cul)
_=plt.plot(count/count.max())
```

```python
def linearmap15(x , xmin , xmax , tmin=0 , tmax=1):
    x/=255
    xmin /= 255
    xmax /= 255
    parameter = (x-xmin) / (xmax-xmin)
    #t = ((x-xmin) / (xmax-xmin)*((x-xmin) / (xmax-xmin)*((x-xmin) / (xmax-xmin)**0.5))) * (tmax-tmin) + tmin
    t = (parameter*(parameter*parameter**0.5)) * ((tmax-tmin) + tmin)
    t*=255
    return t
```
